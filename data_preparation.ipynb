```python
import os
import shutil
import random
```

```python
hr_path = '/home/orelit/Projects -Sathsara/Planigo/data/Wine'
lr_path = '/home/orelit/Projects -Sathsara/Super Resolution and GANs/'
test_data_folder = 'test_images_for_wine'
lr_folder = 'lr'
hr_folder = 'hr'
```

```python
for top_index,folder in enumerate(os.listdir(os.path.join(lr_path,test_data_folder))):
    if folder in os.listdir(os.path.join(hr_path)):
        image_count = 0
        hr_image_list = os.listdir(os.path.join(hr_path,folder))
        for index,img in enumerate(os.listdir(os.path.join(lr_path,test_data_folder,folder))):
            image_count = image_count +1
            shutil.copy(os.path.join(lr_path,test_data_folder,folder,img),os.path.join(lr_path,lr_folder,
                                                                            '{}_{}.jpg'.format(top_index,index)))
        hr_images_to_copy = random.sample(hr_image_list, image_count)
        for index,img in enumerate(hr_images_to_copy):
            print(img)
            shutil.copy(os.path.join(hr_path,folder,img),os.path.join(lr_path,hr_folder,
                                                                      '{}_{}.jpg'.format(top_index,index)))
    else:
        print('Folder {} Not found'.format(folder))
```
